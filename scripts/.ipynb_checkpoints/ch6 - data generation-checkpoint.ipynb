# Household Price Perception

```python
hh_price = hh[['HOUSEID','WTHHFIN','PRICE']]
hh_price = hh_price[hh_price.PRICE.isin([1,2,3,4,5])]

hh_price['price'] = hh_price['PRICE'].map({1:"Strongly agree",2:"Agree",3:"Neither Agree or Disagree",4:"Disagree",5:"Strongly disagree"})
hh_price['price'] = pd.Categorical(hh_price['price'], categories = ["Strongly agree","Agree","Neither Agree or Disagree","Disagree","Strongly disagree"])
hh_price = pd.get_dummies(hh_price,columns=['price'], prefix='', prefix_sep='')
%store hh_price
hh_price.head()
```

# Household AFV type

```python
# household-vehicle by fuel type and -> hybrid type if is alternative fuel
df = veh[['HOUSEID','FUELTYPE','HFUEL']] # some households do not have vehicles

maplist = {(3,):'Alternative Fuels',(2,):'Diesel',(1,):'Gas'} # ,(97,):'Other Fuel'
d = {k: v for tup, v in maplist.items() for k in tup}
df['FUEL'] = df['FUELTYPE'].map(d) # make sure negative values are dropped

maplist = {(3,2):'BEV/PHEV',(4,):'HEV'} # remove other ->,(1,97):'Other'
d = {k: v for tup, v in maplist.items() for k in tup}
df['H'] = df['HFUEL'].map(d)

df = pd.get_dummies(df,columns=['FUEL','H'], prefix='', prefix_sep='')

hh_veh_index = df.groupby(['HOUSEID']).sum().reset_index()

# calculate total number of vehicles
tol_veh = df.groupby(['HOUSEID']).agg({'HOUSEID':'count'}).rename(columns={'HOUSEID':'TOLVEH'})

hh_veh_index = pd.merge(hh_veh_index,tol_veh,on='HOUSEID')

# make sure # hh does not change
print(len(hh))
hh_veh_index = pd.merge(hh[['HOUSEID','WTHHFIN']], hh_veh_index,on='HOUSEID',how='left').fillna(0)
print(len(hh_veh_index))

# NYS region define
state_categoty = hh[['HOUSEID','HHSTATE','WTHHFIN']]
# state_categoty['ISNYS'] = state_categoty['HHSTATE'].map({'NY':"NYS (whole)"}).fillna("Rest of US")
state_categoty['ISNYS'] = 'Rest of US'
state_categoty.loc[state_categoty['HOUSEID'].isin(NYShouseholds),'ISNYS'] = 'NYS Statewide'
state_categoty.loc[state_categoty['HOUSEID'].isin(NYChouseholds),'ISNYC'] = 'NYC'
# state_categoty.loc[state_categoty['HOUSEID'].isin(restNYShouseholds),'ISNYC'] = 'Rest of NYS'
state_categoty.loc[state_categoty['HOUSEID'].isin(NYMTChouseholds),'URBAN'] = 'NYMTC'
state_categoty.loc[state_categoty['HOUSEID'].isin(Urbanhouseholds),'URBAN'] = 'Other Urban (excludes NYMTC)'
state_categoty.loc[state_categoty['HOUSEID'].isin(NonUrbanhouseholds),'URBAN'] = 'Non-Urban Areas'
hh_veh_index = pd.merge(state_categoty[['HOUSEID','ISNYS','ISNYC','URBAN']], hh_veh_index,on='HOUSEID',how='left')
```

```python
# AFV & vehicle ownership
hh_veh_index['AFV adoption'] = ''
hh_veh_index.loc[(hh_veh_index['Alternative Fuels']==0) & (hh_veh_index['TOLVEH']>0),'AFV adoption'] = 'Vehicle, no AFV'
hh_veh_index.loc[hh_veh_index['Alternative Fuels']>0,'AFV adoption'] = 'Vehicle, 1+ AFV' 
hh_veh_index.loc[hh_veh_index['TOLVEH']==0,'AFV adoption'] = 'Zero vehicles'

# low-income indicator
hh_veh_index = pd.merge(hh_veh_index,lowinc_id[['HOUSEID','VLOWINC']],how='left',on='HOUSEID')

%store hh_veh_index
hh_veh_index.head()
```
